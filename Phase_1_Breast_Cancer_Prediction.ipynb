Encoding Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
label_encoding=LabelEncoder()
df['diagnosis']=label_encoding.fit_transform(df['diagnosis'])
```

Finding Outliers Using Boxplot

```python
plt.figure(figsize=(20,15))
plotnumber=1
for col in df.columns:
  if plotnumber <= 25:
    ax=plt.subplot(5,5,plotnumber)
    sns.boxplot(df[col])
    plt.xlabel(col, fontsize=15)
  plotnumber+=1
plt.tight_layout()
plt.show()
```

Splitting Data Variable and Data as x and y

```python
x=df.drop(['diagnosis'],axis=1)
y=df.diagnosis
```

Splitting Data for Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

Standard Scaler for Normal Distribution Data

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler()
x_train=standard_scaler.fit_transform(x_train)
x_test=standard_scaler.transform(x_test)
```

Accuracy Score

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
bagging = BaggingClassifier(KNeighborsClassifier(),
                            max_samples=0.5, max_features=0.5)
bagging.fit(x_train,y_train)
y_pred=bagging.predict(x_test)
print("Accuracy of Bagging classifier:",accuracy_score(y_test,y_pred))
```

Classification Report

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
report=classification_report(y_test,y_pred)
print(report)
```

Confusion Matrix

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

Visualization of Confusion Matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Class 0', 'Class 1'],
            yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
